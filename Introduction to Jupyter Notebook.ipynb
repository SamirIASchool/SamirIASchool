<h2 style="color:red; font-size:30px">1- What's Jupyter Notebook?</h2>

Jupyter notebook is the most efficient way to write code and add visualtiation in the same user interface. It runs in a browser and can be shared quickly.
<br>If you are oriented towards the field of Data, this is the most recommended way. Otherwise, if you are a python application developer, you will need to debug your code and therefore it is recommended to go through IDEs such as pyCharm, visual Code studio, Spyder.
<br>To resume: we use Jupyter Notebook to:
- share a visualization, with or without interactivity.
- document a process with code
- document a library or a module live.
- share code and data in general

<br>A notebook is a collection of `cells`, that can contain different content(Code,Titles, Images, visualisation...). To run a cell: push the start triangle in the menu or type **SHIFT + ENTER/RETURN**
![](../../img/shiftenter.jpg)
<br> For each cell, you have to choice the type (code, markdown, raw text or title) just after the "restar the kernel" button
